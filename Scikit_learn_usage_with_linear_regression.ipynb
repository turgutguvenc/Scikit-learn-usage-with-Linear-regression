## What is the Scikit-Learn and how to use it ?


```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
# attribute matrix(X)(numpySeries or pandas DataFrame)
# target variable (y)(dependent variable)
```

```python
X_iris = iris.drop('species', axis=1) # attribute matrix(X)
X_iris.shape
```

```python
y_iris = iris['species'] # target variable (y)(dependent variable)
y_iris.shape
```

### model operations

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rng = np.random.RandomState(42) # we fix the data
```

```python
# we produce random independent variable and dependent variable
x = 10*rng.rand(50)
y = 2*x-1+rng.randn(50)
plt.scatter(x,y);
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(fit_intercept=True) # model sampling OOP
model # model sampling
```

```python
X = x[:,np.newaxis] #two dimensional attribute matrix
X.shape
```

```python
model.fit(X,y) # model have created
```

```python
model.coef_ # coefficient of X
```

```python
model.intercept_ #intercept
```

```python
x_fit = np.linspace(-1,11,50) #we divide 50 point, between 1, 11 
X_fit = x_fit[:,np.newaxis]  # we set this values to one dimensional array
y_fit =model.predict(X_fit) # we predicted this values
```

```python
plt.scatter(x,y)
plt.plot(x_fit,y_fit)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_iris, y_iris, random_state=1)
# we split the dataset into train and test
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB() # OOP sampling
```

```python
model.fit(X_train, y_train) # we set the model
```

```python
y_model = model.predict(X_test) # we predict the X_test values
```

```python
from sklearn.metrics import accuracy_score
# we import this to find out the model accuracy proporton
```

```python
accuracy_score(y_test, y_model)
# accuracy proportion
# Using this model, we can predict a new iris flower tag 97% accurately.
```
